```python
import numpy as np
import pandas as pd
```

```python
zip_tract = pd.read_csv("datasets/zip_tract.csv")
business_licenses = pd.read_csv("datasets/business_licenses.csv")
```

```python
business_licenses['ZIP CODE'] = business_licenses['ZIP CODE'].fillna(0)
business_licenses['ZIP CODE'] = business_licenses['ZIP CODE'].astype(int)
business_licenses = business_licenses.rename(columns={"ZIP CODE": 'ZIP'})
```

```python
business_licenses_tract = pd.merge(business_licenses, zip_tract, on='ZIP')
business_licenses_tract = business_licenses_tract.drop_duplicates(subset=['LEGAL NAME'])
business_licenses_count = business_licenses_tract.groupby(['TRACT'])['TRACT'].count()
```

```python
business_licenses_count.to_csv("datasets/business_licenses_count.csv")  
```

```python
restaurants_url = 'https://raw.githubusercontent.com/yukichong1998/perceptrons/main/data/Food_Inspections_cleaned.csv?token=GHSAT0AAAAAABQM57II23JQE4CJW4SRECFWYULX6EQ'
```

```python
restaurants = pd.read_csv(restaurants_url)
restaurants = restaurants.rename(columns={"Zip": 'ZIP'})
restaurants_tract = pd.merge(restaurants, zip_tract, on='ZIP')
restaurants_tract = restaurants_tract.drop_duplicates(subset=['DBA Name', 'Address'])
restaurants_count = restaurants_tract.groupby(['TRACT'])['TRACT'].count()
```

```python
restaurants_count.columns = ['tract', 'restaurants_count']
restaurants_count.to_csv("datasets/restaurants_count.csv")  
```
